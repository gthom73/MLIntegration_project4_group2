```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.cm as cm

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Determine global price range for consistent color mapping across all plots, adjusted for thousands
global_min = zip_code_averages['Average_Price'].min() / 1000
global_max = zip_code_averages['Average_Price'].max() / 1000

# Create a single column of plots with increased width for better readability
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(18, 24))

# Define a colormap
cmap = sns.color_palette("viridis", as_cmap=True)

cities = ['Kansas City', 'Topeka', 'Denver', 'Dallas']
for ax, city in zip(axes, cities):
    # Filter the data for each city
    city_data = zip_code_averages[zip_code_averages['city'] == city]
    city_data = city_data.sort_values(by='Average_Price', ascending=False)

    # Calculate the city-wide average price for the horizontal line, adjusted for thousands
    city_avg_price = city_data['Average_Price'].mean() / 1000

    # Normalize the average price for color mapping across all data, adjusted for thousands
    norm = plt.Normalize(global_min, global_max)

    # Create the bar plot with price divided by 1000 for scale adjustment
    bars = sns.barplot(x='zip_code', y=city_data['Average_Price'] / 1000, data=city_data, ax=ax,
                       palette=cmap(norm(city_data['Average_Price'] / 1000)))

    # Draw a horizontal line for the city-wide average price
    ax.axhline(y=city_avg_price, color='red', linestyle='--', label=f'City Avg: ${city_avg_price:.2f}k')

    # Set plot details
    ax.set_title(f'Average Sale Price by Zip Code in {city}')
    ax.set_xlabel('Zip Code')
    ax.set_ylabel('Average Price ($, thousands)')
    ax.tick_params(axis='x', rotation=90, labelsize=10)  # Rotate to vertical for better legibility

    # Add legend to show what the line represents
    ax.legend()

    # Create a colorbar with more detailed description
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar = fig.colorbar(sm, ax=ax, orientation='vertical', fraction=0.02, pad=0.05)
    cbar.set_label('Average Sale Price ($, thousands)', size=12)
    cbar.ax.tick_params(labelsize=10)

plt.tight_layout()
plt.show()

```
